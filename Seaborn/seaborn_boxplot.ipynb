```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Write a program to draw box plot of life expectancy by continent for a data set given in a url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapmind
er-FiveYearData.csv

```python
def main():
    url="https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"

    df = pd.read_csv(url)
    print(df.head())

    sns.boxplot(data=df,x='lifeExp',y='continent',color='green')
    plt.show()


if __name__ == "__main__":
    main()
```

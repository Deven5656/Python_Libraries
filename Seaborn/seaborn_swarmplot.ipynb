```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def main():

    df = sns.load_dataset('tips').head(50)
    print(df.head(10))

    # sns.stripplot(data=df,x='total_bill',y='size',native_scale=True,color='green',size=5)

    sns.swarmplot(data=df,x='total_bill',y='size',native_scale=True,color='green',size=10)
    plt.show()


if __name__ == "__main__":
    main()
```

## 1. Write a Python programming to create a pie chart of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def main():
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.title('Bar Chart')

    plt.pie(popularity,labels=programming_languages)
    
    plt.show()

if __name__ == "__main__":
    main()
```

## 2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def main():
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.title('Bar Chart')

    plt.pie(popularity,labels=programming_languages,autopct='%1.1f%%')
    
    plt.show()

if __name__ == "__main__":
    main()
```

## 3. Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.
Sample data:<br>
medal.csv<br>
country,gold_medal<br>
United States,46<br>
Great Britain,27<br>
China,26<br>
Russia,19<br>
Germany,17


```python
def main():
    df=pd.read_csv('medal.csv')
    print(df)

    tp={'fontsize': 12, 'fontweight': 'bold'}
    wp = {
    'edgecolor': 'black',  # Color of the wedge border
    'linewidth': 1,       # Width of the wedge border
    'linestyle': 'dashed' # Style of the wedge border line
    }

    plt.figure(figsize=(10,10))

    plt.title('Bar Chart')

    w,t,auto=plt.pie(df['gold_medal'],labels=df['country'],explode=[0.2,0,0.1,0,0],startangle=180,
                     shadow=True,autopct='%1.1f%%',textprops=tp,wedgeprops=wp)
    

    for text,wedge,autop in zip(t,w,auto):
        text.set_color(wedge.get_facecolor())
        autop.set_color('w')
    
    plt.legend(title="Countries",loc='upper left')
    plt.show()

if __name__ == "__main__":
    main()
```

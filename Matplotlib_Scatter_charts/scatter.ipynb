```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other

```python
def main():
    x_points=np.random.randint(100,size=50)
    y_points=np.random.randint(100,size=50)

    plt.title('Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    plt.scatter(x_points,y_points,marker='o',edgecolors='red',color='cyan',sizes=np.random.randint(100,size=20))
    
    plt.show()

if __name__ == "__main__":
    main()
```

## 2. Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.

```python
def main():
    x_points=np.random.randint(100,size=50)
    y_points=np.random.randint(100,size=50)

    plt.title('Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    plt.scatter(x_points,y_points,marker='o',edgecolors='red',sizes=np.random.randint(100,size=20),facecolor='None')
    
    plt.show()

if __name__ == "__main__":
    main()
```

## 3. Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes.

```python
def main():
    x_points=np.random.randint(100,size=40)
    y_points=np.random.randint(100,size=40)
    colors=np.random.randint(100,size=40)
    size=np.random.randint(30,100,size=15)

    plt.title('Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    plt.scatter(x_points,y_points,marker='o',c=colors,edgecolors='k',sizes=size,facecolor='None',cmap='Greens_r')
    
    plt.show()

if __name__ == "__main__":
    main()
```

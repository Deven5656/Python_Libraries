```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
def main():
    x_points = np.arange(5)
    y_points = np.arange(5)

    plt.title('1st Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")

    plt.plot(x_points,y_points)
    plt.show()

if __name__ == "__main__":
    main()
```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
def main():

    data=np.loadtxt('test.txt',delimiter=' ')
    print(data)

    x_points = data[:,0]
    y_points = data[:,1]

    plt.title('Test Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    # marker='o',color=red',linestyle=''dotted'

    plt.plot(x_points,y_points,'*',c='r',ls=":")
    plt.show()

if __name__ == "__main__":
    main()
```

## 4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):<br>
Date,Open,High,Low,Close<br>
03-10-16,774.25,776.065002,769.5,772.559998<br>
04-10-16,776.030029,778.710022,772.890015,776.429993<br>
05-10-16,779.309998,782.070007,775.650024,776.469971<br>
06-10-16,779,780.47998,775.539978,776.859985<br>
07-10-16,779.659973,779.659973,770.75,775.080017

```python
def main():

    df=pd.read_csv('fdata.csv')
    print(df)

    plt.figure(figsize=(12,8))


    # plt.xlabel("Date")
    # plt.ylabel("Open")
    plt.subplot(2,2,1)
    plt.plot(df['Date'],df['Open'],label='Open',c='r')
    plt.legend()

    # plt.xlabel("Date")
    # plt.ylabel("Close")
    plt.subplot(2,2,4)
    plt.plot(df['Date'],df['Close'],label='Close')
    plt.legend()

    # plt.xlabel("Date")
    # plt.ylabel("High")
    plt.subplot(2,2,2)
    plt.plot(df['Date'],df['High'],label='High',c='g')
    plt.legend()

   
    # plt.xlabel("Date")
    # plt.ylabel("Low")
    plt.subplot(2,2,3)
    plt.plot(df['Date'],df['Low'],label='Low',c='#000000')
    plt.legend()

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

## 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

```python
def main():

    df=pd.read_csv('fdata.csv')

    plt.figure(figsize=(10,6))

    plt.plot(df['Date'],df['Open'],label='Open',c='r',ls=':')
    plt.plot(df['Date'],df['Close'],label='Close',ls="--")
    plt.plot(df['Date'],df['High'],label='High',c='g',ls='-.')
    plt.plot(df['Date'],df['Low'],label='Low',c='#000000')

    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

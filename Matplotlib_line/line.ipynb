```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
def main():
    x_points = np.arange(5)
    y_points = np.arange(5)

    plt.title('1st Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")

    plt.plot(x_points,y_points)
    plt.show()

if __name__ == "__main__":
    main()
```

## 2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title

```python
def main():
    x_points = [2,4,6,8]
    y_points = [4,2,8,6]

    plt.title('2nd Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")

    plt.plot(x_points,y_points)
    plt.show()

if __name__ == "__main__":
    main()
```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
def main():

    data=np.loadtxt('test.txt',delimiter=' ')
    print(data)

    x_points = data[:,0]
    y_points = data[:,1]

    plt.title('Test Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    # marker='o',color=red',linestyle=''dotted'

    plt.plot(x_points,y_points,'*',c='r',ls=":")
    plt.show()

if __name__ == "__main__":
    main()
```

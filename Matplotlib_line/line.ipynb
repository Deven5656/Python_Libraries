## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
def main():

    data=np.loadtxt('test.txt',delimiter=' ')
    print(data)

    x_points = data[:,0]
    y_points = data[:,1]

    plt.title('Test Plot')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    # marker='o',color=red',linestyle=''dotted'

    plt.plot(x_points,y_points,'*',c='r',ls=":")
    plt.show()

if __name__ == "__main__":
    main()
```

## 9. Write a Python program to display the current axis limits values and set new axis values

```python
def main():
    x = [1, 2, 3, 4, 5]
    y = [2, 3, 5, 7, 11]

    # original axis limits
    plt.subplot(1, 2, 1)
    plt.plot(x, y)

    current_xlim = plt.xlim()
    current_ylim = plt.ylim()
    print("Current X-axis limits:", current_xlim)
    print("Current Y-axis limits:", current_ylim)

    plt.title("Original Limits")
    plt.xlabel("X - Axis")
    plt.ylabel("Y - Axis")

    # new axis limits
    plt.subplot(1, 2, 2)
    plt.plot(x, y)

    plt.xlim(2, 5)
    plt.ylim(0, 10)
    new_xlim = plt.xlim()
    new_ylim = plt.ylim()
    print("New X-axis limits:", new_xlim)
    print("New Y-axis limits:", new_ylim)

    plt.title("Updated Limits")
    plt.xlabel("X - Axis")
    plt.ylabel("Y - Axis")

    # Show both subplots
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    main()
```

## 11. Write a Python program to plot several lines with different format styles in one command using arrays.

```python

x = [10,20,30,40]

y1 = [1,2,3,4]          
y2 = [4,5,6,7]                
y3 = [8,9,10,11]         
y4 = [12,13,14,15]         


formats = ['r--', 'g-.', 'b:', 'k-']  


ys = [y1, y2, y3, y4]


for y, fmt in zip(ys, formats):
    plt.plot(x, y, fmt)


plt.title('Multiple Lines with Different Formats Using Array')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.legend(['1st', '2nd', '3rd', '4th'])

plt.show()
```

## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.

## 7. Write a Python programming to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def main():
    x_points = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    y_points = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    positions = [2,3,4,1,6,5]

    plt.figure(figsize=(10, 6))
    bars = plt.bar(positions, y_points, color='skyblue', edgecolor='black')

    plt.bar_label(bars, labels=y_points, padding=5)

    plt.xticks(positions, x_points, rotation=45)
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    plt.title('Popularity of Programming Languages')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

## 8. Write a Python programming to display a bar chart of the popularity of programming Languages. Select the width of each bar and their positions.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
def main():
    x_points = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    y_points = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    widths = np.array([(np.random.random()) for i in range(len(x_points))])
    positions = [2,3,4,1,6,5]

    plt.figure(figsize=(10, 6))
    bars = plt.bar(positions, y_points, color='skyblue', edgecolor='black',width=widths)

    plt.bar_label(bars, labels=y_points, padding=5)

    plt.xticks(positions, x_points, rotation=45)
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    plt.title('Popularity of Programming Languages')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

## 9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def main():
    x_points = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    y_points = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
   

    plt.figure(figsize=(10, 6))
    bars = plt.bar(x_points, y_points, color='skyblue', edgecolor='black')
    plt.subplots_adjust(bottom=0.2)

    plt.bar_label(bars, labels=y_points, padding=5)

    plt.ylim(0,25)

    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    plt.title('Popularity of Programming Languages')

    plt.show()

if __name__ == "__main__":
    main()
```

## 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:<br>
Means (men) = (22, 30, 35, 35, 26)<br>
Means (women) = (25, 32, 30, 35, 29)

```python
def main():
    points_men = [22,30,35,35,26]
    points_women = [25,32,30,35,29]
    x=np.arange(len(points_men))
   
    width=0.25

    plt.figure(figsize=(10, 6))
    bar1=plt.bar(x, points_men, color='skyblue', edgecolor='black',width=0.25)
    bar2=plt.bar(x+width , points_women, color='hotpink', edgecolor='black',width=0.25)

    plt.bar_label(bar1,labels=['men' for _ in range(len(points_men))],label_type='center',rotation=75)
    plt.bar_label(bar2,labels=['women' for _ in range(len(points_women))],label_type='center',rotation=75)


    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    plt.title('Popularity of Programming Languages')
    plt.xticks(x + width/2 ,['Group A','Group B','Group C','Group D','Group E'])

    plt.show()

if __name__ == "__main__":
    main()
```
